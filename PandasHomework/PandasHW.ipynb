```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# name the csv files
students_file = 'Resources/students_complete copy.csv'
schools_file = 'Resources/schools_complete copy.csv'
```

```python
# create dataframes
df_students = pd.read_csv(students_file)
df_schools = pd.read_csv(schools_file)
```

```python
# District Summary

# 1) Total Schools
total_schools = df_schools['school_name'].value_counts()
total_schools = total_schools.sum()

# 2) Total Students
total_students = df_students['student_name'].value_counts()
total_students = total_students.sum()

# 3) Total Budget
total_budget = df_schools['budget']
total_budget = total_budget.sum()

# 4) Average Math Score
total_math = df_students['math_score']
average_math = round(total_math.mean(), 2)

# 5) Average Reading Score
total_reading = df_students['reading_score']
average_reading = round(total_reading.mean(), 2)

# 6) % Passing Math
passing_math = total_math >= 70
passing_math = passing_math.sum()
percent_passing_math = round(passing_math / total_students, 4)*100

# 7) % Passing Reading
passing_reading = total_reading >= 70
passing_reading = passing_reading.sum()
percent_passing_reading = round(passing_reading / total_students, 4)*100

# 8) # Overall Passing Rate (Average of the above two)
passing_rate = round((passing_math + passing_reading) / (total_students*2), 4)*100
passing_rate

# 9) Create a high level snapshot (in table form) of the district's key metrics
district_summary = pd.DataFrame(
    {'Total Students': [total_students],
     'Total Budget': '$' + ' ' + str(total_budget),
     'Average Math Score': '%' + ' ' + str(average_math),
     'Average Reading Score': '%' + ' ' + str(average_reading),
     'Percent Passing Math': '%' + ' ' + str(percent_passing_math),
     'Percent Passing Reading': '%' + ' ' + str(percent_passing_reading),
     'Overall Passing Rate': '%' + ' ' + str(passing_rate)
    })

district_summary

```
