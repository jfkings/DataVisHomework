```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
business_csv_file = "Resources/business_licenses.csv"
business_df = pd.read_csv(business_csv_file)
business_df
```

```python
business_clean_df = business_df[['LEGAL NAME', 'ADDRESS', 'POLICE DISTRICT', 'APPLICATION REQUIREMENTS COMPLETE']]
business_clean_df
```

```python
business_clean_df = business_clean_df.rename(columns={
    'LEGAL NAME':'legal_name',
    'ADDRESS':'address',
    'POLICE DISTRICT':'police_district',
    'APPLICATION REQUIREMENTS COMPLETE': 'paid_date'
})
business_clean_df
```

```python
crime_csv_file = 'Resources/Chicago_Crimes_2012_to_2017.csv'
crime_df = pd.read_csv(crime_csv_file)
crime_df
```

```python
crime_clean_df = crime_df[['Date', 'Primary Type', 'Description', 'District', 'Arrest']]
crime_clean_df
```
